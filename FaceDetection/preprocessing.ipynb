```python
# Creates training.csv and testing.csv 
from PIL import Image
import numpy as np
import pandas as pd
import random
import os
```

```python
train_fp = "train_DETg9GD/Train/"
test_fp = "test_Bh8pGW3/Test/"
```

```python
train_filenames = os.listdir(train_fp) 
test_filenames = os.listdir(test_fp)
```

```python
def extractPixels(filenames,mode): # mode = train(0)/test(1)
    finalDB = dict()
    if mode:
        maindir = test_fp
    else:
        maindir = train_fp
    
    for filename in filenames:
        fp = maindir + filename
        img = np.array(Image.open(fp).convert("L").resize([28,28]))
        pixels = img.flatten() # 784
        #print(np.shape(pixels))

        finalDB[filename] = pixels
    
    return(finalDB)
        
```

```python
#trainingDB = extractPixels(train_filenames,0)
testingDB = extractPixels(test_filenames,1)
```

```python
training_df = pd.DataFrame.from_dict(trainingDB).T
```

```python
testing_df = pd.DataFrame.from_dict(testingDB).T
```
